```python
from tkinter import * 
from tkcalendar import DateEntry
from tkinter import messagebox



master = Tk()

canvas = Canvas(master, height=450, width=750)
canvas.pack()

frame_ust = Frame(master, bg='#E04BA6')
frame_ust.place(relx=0.1, rely=0.1, relwidth=0.8, relheight=0.1)

frame_sol_alt = Frame(master, bg='#E04BA6')
frame_sol_alt.place(relx=0.1, rely=0.22, relwidth=0.23, relheight=0.5)


frame_sag_alt = Frame(master, bg='#E04BA6')
frame_sag_alt.place(relx=0.35, rely=0.22, relwidth=0.55, relheight=0.5)

gunluk_giris= Label(frame_ust, bg='#E04BA6', text ='Günlük Giriş(TİP):', font='Verdana 12 bold')
gunluk_giris.pack(padx=10, pady=10, side=LEFT)

gunluk_giris_opsiyon = StringVar(frame_ust)
gunluk_giris_opsiyon.set("\t")

gunluk_giris_ac_menu = OptionMenu(frame_ust, gunluk_giris_opsiyon, "1000", "500", "100")
gunluk_giris_ac_menu.pack(padx=8, pady=8, side=LEFT)



gunluk_tarih_ac_menu = DateEntry(frame_ust, widht=12, backgraund='red',foreground='black', borderwidth=1, locale="de_DE")
gunluk_tarih_ac_menu._top_cal.overrideredirect(False)
gunluk_tarih_ac_menu.pack(padx=10, pady=10, side=RIGHT)

gunluk_giris_tarih = Label(frame_ust, bg='#E04BA6', text="Tarih :", font='Verdana 12 bold')
gunluk_giris_tarih.pack(padx=10, pady=10, side=RIGHT)

#####################################################


Label(frame_sol_alt, text="Ürün Stok Kodları", bg="#E04BA6", font="Verdana 12 bold").pack(padx=10, pady=5, anchor=NW)

var = IntVar()

R1 = Radiobutton(frame_sol_alt, text="Type121415", variable = var, value=1, bg="#FFFF00",font="Arial 10")
R1.pack(padx=5, pady=5, anchor=NW)

R2 = Radiobutton(frame_sol_alt, text="Type121416", variable = var, value=2, bg="#FFFF00",font="Arial 10")
R2.pack(padx=5, pady=5, anchor=NW)

R3 = Radiobutton(frame_sol_alt, text="Type121417", variable = var, value=3, bg="#FFFF00",font="Arial 10")
R3.pack(padx=5, pady=5, anchor=NW)

Label(frame_sol_alt, text="Hata Bildirim Yöntemi", bg="#E04BA6", font="Verdana 8 bold").pack(padx=10, pady=4, anchor=W)

var1 = IntVar()

C1 = Checkbutton(frame_sol_alt, text='Email', variable=var1, onvalue=1, offvalue=0,  bg="#FFFF00", font="Verdana 8 bold")
C1.pack(padx=10, pady=2, anchor=NW)

var2 = IntVar()

C2 = Checkbutton(frame_sol_alt, text='Dilekçe', variable=var2, onvalue=1, offvalue=0,  bg="#FFFF00", font="Verdana 8 bold")
C2.pack(padx=10, pady=2, anchor=NW)

#####################################################

def gonder():
    son_mesaj = ""
    try:
        if var.get():
            if var.get() == 1:
                son_mesaj += "Veriniz Basari İle Kaydedildi."
            
                tip = gunluk_giris_opsiyon.get() if gunluk_giris_opsiyon.get() =='' else "Genel"
                tarih = gunluk_tarih_ac_menu.get()
                mesaj = metin_alani.get("1.0", "end")
            
                with open ("Stok.txt", "w") as dosya:
                    dosya.write(
                    '{} adetinde,{} tarihinde '.format(
                        tip,
                        tarih,
                        mesaj
                    
                    ))
            
                    dosya.close()
                  
            
            elif var.get() == 2:
                son_mesaj += "E posta yolu ile gönderilmiştir."
            
                messagebox.showinfo("Basarili Islem", son_mesaj)
            
        else:
            son_mesaj += "Mesaji eklediğinizden emin olun."
            messagebox.showinfo("Mesaji doldurmadiniz.", son_mesaj)
    except:
        son_mesaj += "İslem Basarisiz"
        messagebox.showerror("Basarisiz", son_mesaj)
    finally:
        master.destroy()


Label(frame_sag_alt, text="Hatalı Ürün Açıklama Mesaji", bg="#E04BA6", font="Verdana 12 bold").pack(padx=10, pady=5, anchor=NW)

metin_alani = Text(frame_sag_alt, height = 9, width=50)
metin_alani.tag_configure('style', foreground = '#FF0000', font=('Verdana', 7, 'bold'))
metin_alani.pack()


karsilama_metni = 'Mesajini buraya gir...'
metin_alani.insert(END, karsilama_metni, 'style')

gonder_Button = Button(frame_sag_alt, text="Gönder", command= gonder)

gonder_Button.pack(anchor=S)


master.mainloop()
```
